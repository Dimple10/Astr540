### Part a

```python
from classy import Class
import matplotlib.pyplot as plt
import numpy as np

cosmo = Class()
cosmo.set({'output':'tCl','lensing':'no'})
cosmo.compute()
cls = cosmo.raw_cl(2500)
ll = cls ['ell'][2:]
clTT = cls ['tt'][2:]
clTT = clTT*ll*(ll+1)/2./np.pi

fig_TT, ax_TT = plt.subplots()
ax_TT.semilogx(ll,clTT,linestyle='-')

cosmo.struct_cleanup()

ax_TT.set_title('Angular Power Spectrum')
ax_TT.set_xlim([2,2500])
ax_TT.set_xlabel(r'$\ell$')
ax_TT.set_ylabel(r'$C_\ell^\mathrm{TT}$')
plt.show()
```

### Part b

```python
cosmo.set({'omega_b':0.02238280*1.10})
cosmo.compute()
cls = cosmo.raw_cl(2500)
ll = cls ['ell'][2:]
b_clTT = cls ['tt'][2:]
b_clTT = b_clTT*ll*(ll+1)/2./np.pi

fig_TT, ax_TT = plt.subplots()
ax_TT.semilogx(ll,b_clTT,linestyle='-', label='Increased')
ax_TT.semilogx(ll,clTT,linestyle='-', color='orange', label='Fiducial')
cosmo.struct_cleanup()

ax_TT.set_title('Baryon density increased by 10%')
ax_TT.set_xlim([2,2500])
ax_TT.set_xlabel(r'$\ell$')
ax_TT.set_ylabel(r'$C_\ell^\mathrm{TT}$')
plt.legend()
plt.show()
```
